```python
# Feature and target separation
features = train.drop("price", axis=1)
target = train["price"]
```

```python
# Perform train-test split
X_train, X_val, y_train, y_val = train_test_split(features, target, test_size=0.2, random_state=42)

# Verify the shape of the splits
(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
# One-hot encode the categorical features
X_train_encoded = pd.get_dummies(X_train, drop_first=True)
X_val_encoded = pd.get_dummies(X_val, drop_first=True)

# Align the columns of the validation set to match the training set
X_val_encoded = X_val_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)

# Impute remaining missing values with the median for numerical features
imputer = SimpleImputer(strategy='median')
X_train_encoded_imputed = imputer.fit_transform(X_train_encoded)
X_val_encoded_imputed = imputer.transform(X_val_encoded)

# Train the Linear Regression model on the imputed, encoded data
linear_model = LinearRegression()
linear_model.fit(X_train_encoded_imputed, y_train)

# Predict on the validation set
y_pred = linear_model.predict(X_val_encoded_imputed)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_val, y_pred))

# Display RMSE
print(f"RMSE: {rmse}")
```

```python
# Example of loading test_df, ensure to load it correctly
# test_df = pd.read_csv('path_to_test_file.csv')  # Make sure to load the test data correctly

# Verify the test_df columns
#print("Columns in test_df before preprocessing:")
#print(test.columns)

# Extract and preserve the 'id' column
test_id = test['id']

# One-hot encode the categorical features in the test dataset
test_df_encoded = pd.get_dummies(test.drop(columns=['id']), drop_first=True)

# Align the columns of the test set to match the training set
test_df_encoded = test_df_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)

# Impute remaining missing values in the test set with the median for numerical features
test_df_imputed = imputer.transform(test_df_encoded)

# Predict on the test dataset using the trained Linear Regression model
test_predictions = linear_model.predict(test_df_imputed)

```

```python
# Prepare the submission DataFrame
submission_df = pd.DataFrame({
    'id': test_id,  # Add the preserved 'id' column back
    'price': test_predictions
})

# Ensure the 'price' is in integer format
submission_df['price'] = submission_df['price'].round().astype(int)



# Display the first few rows of the submission file
print(submission_df.head())
```

```python
# Save the submission DataFrame to a CSV file
submission_file_path = 'submission.csv'
submission_df.to_csv(submission_file_path, index=False)
```
